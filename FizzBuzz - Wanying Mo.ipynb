# FizzBuzz
To solve the problem of 'FizzBuzz'.

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print ('FizzBuzz')
    elif i % 3 == 0 and i % 7 == 0:
        print ('FizzQizz')
    elif i % 5 == 0 and i % 7 == 0:
        print ('BuzzQizz')
    elif i % 3 == 0 and i % 5 and i % 7 == 0:
        print ('FizzBuzzQizz')
    elif i % 3 == 0:
        print ('Fizz')
    elif i % 5 == 0:
        print ('Buzz')
    elif i % 7 == 0:
        print ('Qizz')
    else:
        print (str(i))
```

# Tic Tac Toe
To find the winner of the game.

```python
def check_board(board):
    n=0
    for x in range(3):
        if board[x][0] == board[x][1] and board[x][1] == board[x][2]:
            print(board[x][0] + ' won')
            n = n+1
        if board[0][x] == board[1][x] and board[1][x] == board[2][x]:
            print(board[0][x] + ' won')
            n = n+1
        if board[0][0] == board[1][1] and board[1][1] == board[2][2]:
            print(board[0][0] + ' won')
            n = n+1
        if board[0][2] == board[1][1] and board[1][1] == board[2][0]:
            print(board[0][2] + ' won')
            n = n+1
        
            if n == 0:
                print('draw')
        
board = [['O', 'X', 'X'],
         ['X', 'X', 'O'],
         ['O', 'X', 'X']]
check_board(board)
```
